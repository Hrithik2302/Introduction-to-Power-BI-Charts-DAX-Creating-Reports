#answers

#ans1. Power BI & Components

Power BI is a Microsoft business analytics tool used to visualize data and share insights.

Power BI Desktop: Create reports, data models, DAX.

Power BI Service: Publish, share dashboards online.

Power BI Mobile: View reports on mobile devices.

Power BI Gateway: Connect on-premise data to cloud.

#ans2. Visual Comparison

Pie vs Donut

Pie: Simple part-to-whole view

Donut: Same as pie with center space
Example: Sales % by Region

Bar vs Column

Bar: Horizontal, good for long names

Column: Vertical, good for time data
Example: Bar → Sales by Sub-Category, Column → Monthly Sales

#ans3. Modeling Concepts

Star vs Snowflake

Star: Simple, faster, preferred in Power BI

Snowflake: Complex, normalized

Primary vs Foreign Key

Primary Key: Unique identifier

Foreign Key: Links tables

Cardinality: Ensures correct data aggregation (1:1, 1:Many).

#ans4. Calculated Column vs Measure

Calculated Column: Row-level, stored

Measure: Aggregate, dynamic

Row Context: Works row by row
Filter Context: Works with filters/slicers

#ans5. Report vs Dashboard

Report: Multi-page, interactive, built in Desktop

Dashboard: Single page, summary, built in Service

#ans8. % of Total Sales (DAX)
Sales % =
DIVIDE(
  SUM(Sales[Sales_Amount]),
  CALCULATE(SUM(Sales[Sales_Amount]), ALL(Sales))
)
